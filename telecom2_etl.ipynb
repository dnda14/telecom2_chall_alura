#### Diccionario de datos

- `customerID`: número de identificación único de cada cliente

- `Partner`: si el cliente tiene o no una pareja
- `Dependents`: si el cliente tiene o no dependientes
- `tenure`: meses de contrato del cliente
- `PhoneService`: suscripción al servicio telefónico
- `MultipleLines`: suscripción a más de una línea telefónica
- `InternetService`: suscripción a un proveedor de internet
- `OnlineSecurity`: suscripción adicional de seguridad en línea
- `OnlineBackup`: suscripción adicional de respaldo en línea
- `DeviceProtection`: suscripción adicional de protección del dispositivo
- `TechSupport`: suscripción adicional de soporte técnico, menor tiempo de espera
- `StreamingTV`: suscripción de televisión por cable
- `StreamingMovies`: suscripción de streaming de películas
- `Contract`: tipo de contrato
- `PaperlessBilling`: si el cliente prefiere recibir la factura en línea
- `PaymentMethod`: forma de pago
- `Charges.Monthly`: total de todos los servicios del cliente por mes
- `Charges.Total`: total gastado por el cliente

## Cantidad de grupo de par de servicios contratados

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

servicios_cols = ['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
df_bool = (df[servicios_cols] == 'Yes').astype(int)

co_matrix = df_bool.T.dot(df_bool)

print(co_matrix)
import numpy as np

np.fill_diagonal(co_matrix.values, 0)

max_frecuencia = co_matrix.max().max()
par_frecuente = co_matrix.stack().idxmax()

print(f"El par más frecuente es {par_frecuente} con {max_frecuencia} clientes.")
plt.figure(figsize=(10, 8))
sns.heatmap(co_matrix, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Matriz de Co-ocurrencia de Servicios')
plt.show()
```

```python

from itertools import combinations

servicios_cols = ['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']

df_bool = (df[servicios_cols] == 'Yes')

todas_las_combinas = list(combinations(servicios_cols, 3))
conteo_trios = {}

for trio in todas_las_combinas:
    frecuencia = df_bool[list(trio)].all(axis=1).sum()
    conteo_trios[trio] = frecuencia

resultados_trios = pd.Series(conteo_trios).sort_values(ascending=False)


top_trios = resultados_trios.head(10)
top_trios.index = [' + '.join(i) for i in top_trios.index]
top_trios = top_trios.reset_index()
top_trios.columns = ['Trío', 'Frecuencia']

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_trios,
    x='Frecuencia',
    y='Trío',
    hue='Trío',      # Para evitar el warning de palette
    palette='viridis',
    legend=False
)

plt.title('Top 10 de Combinaciones de 3 Servicios')
plt.xlabel('Cantidad de Clientes')
plt.ylabel('Combinación de Servicios')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from imblearn.over_sampling import SMOTE

sns.set_style('whitegrid')
%matplotlib inline
```

```python
df.drop('customerID', axis=1, inplace=True)
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')
df.dropna(inplace=True)
df.shape
```

```python

categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

categorical_cols.remove('Churn')

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df_encoded.head()
```

```python
df_encoded['Churn'] = df_encoded['Churn'].map({'Yes': 1, 'No': 0})

df_encoded['Churn'].value_counts()
```

```python
churn_prop = df_encoded['Churn'].value_counts(normalize=True)
print("Proporción de cancelaciones:\n", churn_prop)

sns.countplot(x='Churn', data=df_encoded)
plt.title('Distribución de Cancelaciones')
plt.show()
```

```python

numeric_df = df_encoded.select_dtypes(include=[np.number])

plt.figure(figsize=(14, 10))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación - Variables Numéricas')
plt.show()

corr_with_churn = corr_matrix['Churn'].sort_values(ascending=False)
print("Correlación de variables con Churn:\n", corr_with_churn)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Churn', y='tenure', data=df_encoded)
plt.title('Distribución de tenure según cancelación')
plt.xlabel('Cancelación (0=No, 1=Sí)')
plt.ylabel('Meses de contrato')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Churn', y='Charges.Total', data=df_encoded)
plt.title('Distribución de cargo total según cancelación')
plt.xlabel('Cancelación (0=No, 1=Sí)')
plt.ylabel('Cargo total ($)')
plt.show()
```

```python
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

print(f"Antes de limpieza: X shape {X.shape}, y shape {y.shape}")
print(f"NaN en y: {y.isnull().sum()}")
print(f"NaN en X (por columna):\n{X.isnull().sum()}")

mask_y = y.notna()
X = X[mask_y]
y = y[mask_y]

mask_x = X.notna().all(axis=1)
X = X[mask_x]
y = y[mask_x]

print(f"Después de limpieza: X shape {X.shape}, y shape {y.shape}")
print(f"NaN restantes en y: {y.isnull().sum()}")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Tamaño del entrenamiento: {X_train.shape}")
print(f"Tamaño de prueba: {X_test.shape}")
print(f"Distribución de Churn en entrenamiento:\n{y_train.value_counts(normalize=True)}")
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("Distribución después de SMOTE:")
print(y_train_resampled.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
```

```python

numeric_features = ['tenure', 'Charges.Monthly', 'Charges.Total']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features)
    ],
    remainder='passthrough'
)
```

```python
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000))
])

lr_pipeline.fit(X_train, y_train)

y_pred_lr = lr_pipeline.predict(X_test)
y_proba_lr = lr_pipeline.predict_proba(X_test)[:, 1]

print("=== REGRESIÓN LOGÍSTICA ===")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_lr))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred_lr))
```

```python
cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión - Regresión Logística')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()

fpr, tpr, _ = roc_curve(y_test, y_proba_lr)
plt.plot(fpr, tpr, label=f'ROC (AUC = {roc_auc_score(y_test, y_proba_lr):.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC - Regresión Logística')
plt.legend()
plt.show()
```

```python
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42, n_estimators=100))
])

rf_pipeline.fit(X_train, y_train)

y_pred_rf = rf_pipeline.predict(X_test)
y_proba_rf = rf_pipeline.predict_proba(X_test)[:, 1]

print("=== RANDOM FOREST ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_rf))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred_rf))
```

```python
cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens')
plt.title('Matriz de Confusión - Random Forest')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
plt.plot(fpr_rf, tpr_rf, label=f'ROC (AUC = {roc_auc_score(y_test, y_proba_rf):.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC - Random Forest')
plt.legend()
plt.show()
```

```python
print("Comparación de modelos:")
print(f"Regresión Logística - Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}, AUC: {roc_auc_score(y_test, y_proba_lr):.4f}")
print(f"Random Forest      - Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}, AUC: {roc_auc_score(y_test, y_proba_rf):.4f}")
```
